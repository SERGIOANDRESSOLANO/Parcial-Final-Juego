```python
import random

def generar_numero_secreto():
    digitos = list("0123456789")
    random.shuffle(digitos)
    return "".join(digitos[0:4])

def jugar_fijas_y_picas():
    numero_secreto = generar_numero_secreto()
    intentos = 0
    print("JUEGO DE FIJAS Y PICAS")
    print("Encuentra el número de 4 dígitos")

    while True:
        intento_usuario = input("Introduce tu cifra (4 dígitos únicos): ")

        if len(intento_usuario) != 4 or not intento_usuario.isdigit():
            print("Introduce un número válido de 4 dígitos.")
            continue

        if len(set(intento_usuario)) != 4:
            print("Por favor, introduce 4 dígitos únicos en tu intento.")
            continue

        intentos += 1
        fijas = 0
        picas = 0

        for i in range(4):
            if intento_usuario[i] == numero_secreto[i]:
                fijas += 1

        secreto_sin_fijas = []
        intento_sin_fijas = []

        for i in range(4):
            if intento_usuario[i] != numero_secreto[i]:
                secreto_sin_fijas.append(numero_secreto[i])
                intento_sin_fijas.append(intento_usuario[i])

        for digito_intento in intento_sin_fijas:
            if digito_intento in secreto_sin_fijas:
                picas += 1
                secreto_sin_fijas.remove(digito_intento)

        print("Intento {}: {} - Fijas: {}, Picas: {}".format(intentos, intento_usuario, fijas, picas))

        if fijas == 4:
            print("Descubriste el número {} en {} intentos.".format(numero_secreto, intentos))
            break

jugar_fijas_y_picas()
```
